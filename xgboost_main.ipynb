# XGBoost
## Introduction and importing data
This is the main file for the XGBoost. Here we call all the necessary functions and train the XGBoost. Let's get started by first importing the data.

```python
import data_utils as du
import learning_utils as lu
import visu_utils as vu
```

```python
data_file = 'kaggle_data/train_data.csv'
labels_file = 'kaggle_data/train_labels.csv'

data, labels = du.Import_Data(data_file, labels_file)
```

## Preprocessing data
Here we remove all features with zero variance, since they're not useful and also because they will cause a division by zero during normalization. Normalization is done here using min-max method. After normalization we shuffle the data, in order for Keras to pick a decent training set and validation set. We do not need to split the data into test and train sets, since Keras has a build in method for doing that during training.

```python
clean_data = du.Remove_Zero_Variance(data)
normalized_features = du.Normalize(clean_data, 'min-max')
shuffled_features, shuffled_labels = du.Shuffle(normalized_features, labels)
shuffled_features.shape
train_features, test_features, train_labels, test_labels = du.Split_Data(shuffled_features, shuffled_labels, 0.33)
```

## Growing the random forest


```python
bst = lu.Learn_XGBoost(train_features,train_labels,test_features,test_labels,max_depth=500,num_round=1000,use_weights=False)
```

```python
pred_labels, pred_labels_score = lu.XGB_Predict(bst,test_features,test_labels)
vu.ROC_curve(test_labels,pred_labels_score, fig_size=(12,10))
```

```python
vu.performance_report(test_labels,pred_labels)
```

```python
bst_balanced = lu.Learn_XGBoost(train_features,train_labels,test_features,test_labels,max_depth=500,num_round=1000,use_weights=True)
```

```python
pred_bal_labels, pred_bal_labels_score = lu.XGB_Predict(bst_balanced,test_features,test_labels)
vu.ROC_curve(test_labels,pred_bal_labels_score, fig_size=(12,10))
vu.performance_report(test_labels,pred_bal_labels)
```

```python
bst_balanced2 = lu.Learn_XGBoost(train_features,train_labels,test_features,test_labels,max_depth=900,num_round=5000,use_weights=True)
```

```python
pred_bal_labels2, pred_bal_labels_score2 = lu.XGB_Predict(bst_balanced2,test_features,test_labels)
vu.ROC_curve(test_labels,pred_bal_labels_score2, fig_size=(12,10))
vu.performance_report(test_labels,pred_bal_labels2)
```
